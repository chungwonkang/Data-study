# 필요한 컬럼 만들기

```python
df.columns
```

할인율을 적용하기 전의 순 가격을 계산해보자

```python
df['net_price'] = df.price * (1- df.discount_ratio)
df['total_net_revenue'] = df.net_price * df.quantity
df['total_revenue'] = df.price * df.quantity
df['total_cost'] = df.cost * df.quantity
df['discount'] = df.discount_ratio * df.price
df['total_discount'] = df.discount * df.quantity
df.head()
```

데이터 프레임을 product_id와 카테고리에 따라서 그룹화 해서 각 그룹별로 total 값을 합산하고 비율을 구해보자

```python
df_agg = df.groupby( by = ['product_id', 'category', 'price'])[[
    'quantity',
    'total_net_revenue',
    'total_revenue',
    'total_cost',
    'total_discount'

    ]].sum().reset_index()
df_agg.head()
```

```python
df_agg['net_profit_ratio'] = 1-(df_agg.total_cost + df_agg.total_discount)/df_agg.total_revenue
```

```python
df_agg['total_net_profit'] = df_agg.total_revenue - df_agg.total_cost - df_agg.total_discount
```

```python
df_agg['discount_ratio'] = df_agg.total_discount / df_agg.total_revenue
```

```python
df_agg['cost_ratio'] = df_agg.total_cost / df_agg.total_revenue
```

```python
df_agg
```

결과가 맞는지 검산해보는 코드를 만들어보자

```python
df_agg.net_profit_ratio + df_agg.cost_ratio + df_agg.discount_ratio
```

```python
df_agg.total_revenue - df_agg.total_cost - df_agg.total_discount - df_agg.total_net_profit
```

# 시각화

버블 차트로 시각화 해보자
버블 차트 - 스캐터 차트의 특수한 형태

```python
fig = px.scatter(
    df_agg
    , x = 'total_revenue'
    , y = 'net_profit_ratio'
    , height = 600
    , width = 800
    , title = '제품 포트폴리오 분석'
    , hover_name = 'product_id'
    , size = 'total_net_profit'
    , color = 'discount_ratio'
)
fig.show()
```

4사분면의 경우에는 할인을 많이 한 11과 할인이 적은 19, 15의 경우가 보인다.
11을 더욱 이익 극대화를 하기 위해서는, 제품 중 가장 큰 폭의 할인을 하고 있기 때문에 매출 볼륨을 조금더 손해를 보더라도 할인 폭을 낮춰서 제품 이익률을 높였을 때 전체적인 이윤을 더 높일 수 있을 것으로 보인다.

다음 19, 15의 경우에는 할인이 적어도 매출이 많이 발생하고 있어, 할인을 한다 해도 매출을 더 올릴 수 있겠지만 할인율이 상당히 안좋기 때문에 전체적인 이윤을 올릴 수 있을지는 알 수 없을 것으로 보인다. 그럼 이 제품의 가격을 올린다면 얼마나 민감하게 영향을 미칠지 모르기 때문에 가격 탄력도에 대한 더 면밀한 분석과 검토가 필요해 보인다.
